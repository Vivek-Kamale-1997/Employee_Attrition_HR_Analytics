```python
# 1] importing all python package libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2] Load the Dataset
filename="D:\Onedrive\Desktop\HR Analytics\Python skill course\EmployeeAttritionProject\employee_attrition.csv"
df=pd.read_csv(filename)
df.head(5)
```

```python
# Visualization settings
plt.style.use("default")
sns.set_theme()
```

```python
#3]  dataset shape(rows,columns)
df.shape
```

```python
# 4] dataset information
df.info()
```

```python
# 5] Checking Missing Values
df.isnull().sum
```

```python
# 6] Attrition Distribution
# count attrition=yes and attrition=no
df['Attrition'].value_counts()
```

```python
# 7] Attrition Rate (Percentage)
df['Attrition'].value_counts(normalize=True) * 100
```

```python
# 8] Attrition Visualization
sns.countplot(x='Attrition', data=df)
plt.title("Employee Attrition Count")
plt.show()
```

```python
# 9] Attrition by Department
pd.crosstab(df['Department'], df['Attrition'])
```

```python
# 10] Attrition % by Department
pd.crosstab(df['Department'], df['Attrition'], normalize='index') * 100
```

```python
# 11] Visualization â€“ Attrition by Department
sns.countplot(x='Department', hue='Attrition', data=df)
plt.title("Attrition by Department")
plt.xticks(rotation=20)
plt.show()
```

```python
# 12] Attrition by Gender
pd.crosstab(df['Gender'], df['Attrition'])
```

```python
# 13] Attrition by Job Satisfaction
pd.crosstab(df['JobSatisfaction'], df['Attrition'])
```

```python
# 14] Monthly Income vs Attrition
# Employees with lower income leave more
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)
plt.title("Monthly Income vs Attrition")
plt.show()
```

```python
# 15] Employee Age Distribution
sns.histplot(df['Age'], bins=20, kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
# 16] Attrition by Work-Life Balance
# Poor work-life balance â†’ higher attrition
pd.crosstab(df['WorkLifeBalance'], df['Attrition'], normalize='index') * 100
```

```python
# 17] Correlation Heatmap
# Shows relationships between income, age, experience
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
